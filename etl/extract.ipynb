```python
from statsbombpy import sb
import pandas as pd
from pathlib import Path
```

```python
RAW_DIR = Path("../data/raw")
RAW_DIR.mkdir(parents=True, exist_ok=True)
```

```python
competitions = sb.competitions()
competitions.head()
```

```python
competitions.to_csv(RAW_DIR / "competitions.csv", index=False)
```

```python
prem_matches = sb.matches(
    competition_id=2 ,
    season_id=27
)

prem_matches
```

```python
prem_matches.to_csv(
    RAW_DIR / "matches_epl_2015_16.csv",
    index=False
)
```

```python
events_epl = []

for match_id in prem_matches["match_id"]:
    df = sb.events(match_id=match_id)
    df["match_id"] = match_id
    events_epl.append(df)

events_epl = pd.concat(events_epl, ignore_index=True)
events_epl.shape
```

```python
events_epl.to_csv(
    RAW_DIR / "events_epl_2015_16.csv",
    index=False
)
```
